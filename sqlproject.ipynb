```python
%%sql 

SELECT category,
      AVG("price") AS avg_price,
      AVG("net") AS avg_net, 
      SUM("daily_sales") AS daily_sales,
      AVG("rating") AS avg_rating
FROM keyterms
WHERE category = 'Beauty & Personal Care'
GROUP BY category; 
```

```python
%%sql 

SELECT category, 
       COUNT(product_name) AS total_number_of_products,
       AVG(price) AS avg_price, 
       SUM(monthly_sales) AS avg_monthly_sales,
       SUM(daily_sales) AS total_daily_sales,
       SUM(monthly_revenue) AS total_monthly_revenue,
       AVG(net) AS avg_net,
       AVG(fees) AS avg_fees
FROM keyterms
GROUP BY category;
```
